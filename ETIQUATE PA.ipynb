```python
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib.units import mm
from reportlab.pdfgen import canvas
from reportlab.lib.utils import ImageReader

def generate_cards(output_file, top_logo, bottom_logo):
    # Page setup
    page_width, page_height = landscape(A4)
    margin_x = 14 * mm   # 1.4 cm
    margin_y = 8 * mm   # 1 cm

    usable_width = page_width - 2 * margin_x
    usable_height = page_height - 2 * margin_y

    cols = 4
    rows = 2
    card_width = usable_width / cols
    card_height = usable_height / rows

    # Load logos
    top_img = ImageReader(top_logo)
    bottom_img = ImageReader(bottom_logo)

    # Create PDF
    c = canvas.Canvas(output_file, pagesize=landscape(A4))

    number = 1
    for n in range(600):  # 600 cartes
        if n % (rows * cols) == 0 and n != 0:
            c.showPage()

        # Position
        page_index = n % (rows * cols)
        row = page_index // cols
        col = page_index % cols

        x = margin_x + col * card_width
        y = margin_y + (rows - 1 - row) * card_height
        x_center = x + card_width / 2   # center of the card

        # Optional: border to help cutting
        #c.rect(x, y, card_width, card_height)

        # ---------- Layout spacing ----------
        blank_bottom = 15 * mm   # 1.5 cm
        gap_logo_num = 14 * mm   # 1.4 cm
        gap_num_phrase = 9 * mm  # 0.9 cm
        gap_phrase_logo = 2 * mm # 0.2 cm

        # Bottom logo
        bottom_height = 12 * mm
        bottom_width = card_width * 0.4
        bottom_y = y + blank_bottom
        c.drawImage(bottom_img,
                    x_center - bottom_width/2,
                    bottom_y,
                    width=bottom_width,
                    height=bottom_height,
                    preserveAspectRatio=True,
                    mask='auto')

        # Phrase "Traité par"
        phrase_y = bottom_y + bottom_height + gap_phrase_logo
        c.setFont("Helvetica", 14)
        c.drawCentredString(x_center, phrase_y, "Traité par")

        # Number
        num_y = phrase_y + gap_num_phrase
        c.setFont("Helvetica-Bold", 50)
        c.drawCentredString(x_center, num_y, f"{number:04d}")

        # Top logo (full width)
        logo_width = card_width * 0.80
        logo_height = card_height * 0.5
        top_y = num_y + gap_logo_num + logo_height
        c.drawImage(top_img,
                    x_center - logo_width/2,
                    top_y - logo_height,
                    width=logo_width,
                    height=logo_height,
                    preserveAspectRatio=True,
                    mask='auto')

        number += 1

    c.save()

     # Example usage:
generate_cards("cartes.pdf", "top_logo.png", "bottom_logo.png")
```
